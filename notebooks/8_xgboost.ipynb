# Part I

## Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import xgboost as xgb
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
```

## Step 2: Load Dataset

```python
data = load_iris()
X = data.data
y = data.target

print("Feature names:", data.feature_names)
print("Target classes:", data.target_names)

print("X shape:", X.shape)
print("y shape:", y.shape)
```

## Step 5: Make Predictions

```python
y_pred = model.predict(X_test)

# Accuracy
print("Accuracy:", accuracy_score(y_test, y_pred))

# Confusion matrix
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred))

# Full classification report
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## Step 2: Helper Functions

```python
# 1. Function to estimate Mean and Variance
def estimateGaussian(X):
    m, n = X.shape
    mu = np.mean(X, axis=0)
    sigma2 = np.var(X, axis=0)
    
    return mu, sigma2


# 2. Function to calculate probability density (Provided in your files)
def multivariateGaussian(X, mu, Sigma2):
    k = mu.size
    
    if Sigma2.ndim == 1:
        Sigma2 = np.diag(Sigma2)
        
    X = X - mu
    p = (2 * np.pi) ** (- k / 2) * np.linalg.det(Sigma2) ** (-0.5) * \
        np.exp(-0.5 * np.sum(np.dot(X, np.linalg.pinv(Sigma2)) * X, axis=1))
    
    return p


# 3. Function to visualize the data
def visualizeFit(X, mu, sigma2):
    X1, X2 = np.meshgrid(np.arange(0, 35.5, 0.5), np.arange(0, 35.5, 0.5))
    
    Z = multivariateGaussian(np.stack([X1.ravel(), X2.ravel()], axis=1), mu, sigma2)
    Z = Z.reshape(X1.shape)
    
    plt.plot(X[:, 0], X[:, 1], 'bx')
    
    if np.all(abs(Z) != np.inf):
        plt.contour(X1, X2, Z, levels=10**(np.arange(-20., 1, 3)), zorder=100)
```

## Step 3: Selecting the Threshold

```python
def selectThreshold(yval, pval):
    bestEpsilon = 0
    bestF1 = 0
    F1 = 0
    
    step = (max(pval) - min(pval)) / 1000
    
    for epsilon in np.arange(min(pval), max(pval), step):
        # Predictions: 1 if anomaly (prob < epsilon), 0 otherwise
        predictions = (pval < epsilon)
        
        # Calculate True Positives, False Positives, False Negatives
        tp = np.sum((predictions == 1) & (yval == 1))
        fp = np.sum((predictions == 1) & (yval == 0))
        fn = np.sum((predictions == 0) & (yval == 1))
        
        # Calculate Precision and Recall
        prec = tp / (tp + fp) if (tp + fp) > 0 else 0
        rec = tp / (tp + fn) if (tp + fn) > 0 else 0
        
        # Calculate F1 Score
        if (prec + rec) > 0:
            F1 = 2 * prec * rec / (prec + rec)
        else:
            F1 = 0
            
        if F1 > bestF1:
            bestF1 = F1
            bestEpsilon = epsilon
            
    return bestEpsilon, bestF1
```

## Step 5: The Cost Function

```python
def cofiCostFunc(params, Y, R, num_users, num_movies, num_features, lambda_=0.0):
    # Unfold the params back into matrices X and Theta
    X = params[:num_movies*num_features].reshape(num_movies, num_features)
    Theta = params[num_movies*num_features:].reshape(num_users, num_features)
    
    # 1. Calculate Error
    # We only look at errors where there was actually a rating (R=1)
    predictions = np.dot(X, Theta.T)
    error = (predictions - Y) * R
    
    # 2. Calculate Cost (J) with Regularization
    J = 0.5 * np.sum(error**2)
    J += (lambda_ / 2) * (np.sum(Theta**2) + np.sum(X**2))
    
    # 3. Calculate Gradients (X_grad and Theta_grad) with Regularization
    X_grad = np.dot(error, Theta) + (lambda_ * X)
    Theta_grad = np.dot(error.T, X) + (lambda_ * Theta)
    
    grad = np.concatenate([X_grad.ravel(), Theta_grad.ravel()])
    return J, grad
```

```python
def normalizeRatings(Y, R):
    m, n = Y.shape
    Ymean = np.zeros(m)
    Ynorm = np.zeros(Y.shape)
    for i in range(m):
        idx = R[i, :] == 1
        if np.any(idx):
            Ymean[i] = np.mean(Y[i, idx])
            Ynorm[i, idx] = Y[i, idx] - Ymean[i]
    return Ynorm, Ymean

# Load the movie data
data = loadmat(EX8MOVIES)
Y, R = data['Y'], data['R']

print(f"Average rating for movie 1 (Toy Story): {np.mean(Y[0, R[0, :] == 1]):.2f} / 5")
```

## Step 7: Training the Model

```python
# 1. Setup helpful variables
num_movies, num_users = Y.shape
num_features = 10 # We assume 10 hidden features (action, romance, etc.)

# 2. Add YOUR own ratings (Optional - simulating a new user)
my_ratings = np.zeros(num_movies)
my_ratings[0] = 4   # Toy Story
my_ratings[97] = 2  # Silence of the Lambs
my_ratings[6] = 3   # Twelve Monkeys
my_ratings[11]= 5   # Usual Suspects
my_ratings[53] = 4  # Outbreak
my_ratings[63] = 5  # Shawshank Redemption
my_ratings[65] = 3  # While You Were Sleeping
my_ratings[68] = 5  # Forrest Gump
my_ratings[182] = 4 # Alien
my_ratings[225] = 5 # Die Hard 2
my_ratings[354] = 5 # Sphere

# Add new user to the data
Y = np.hstack([my_ratings[:, None], Y])
R = np.hstack([(my_ratings > 0)[:, None], R])

# Normalize
Ynorm, Ymean = normalizeRatings(Y, R)
num_users = Y.shape[1]

# 3. Initialize Random Parameters
X = np.random.randn(num_movies, num_features)
Theta = np.random.randn(num_users, num_features)
initial_parameters = np.concatenate([X.ravel(), Theta.ravel()])

# 4. Train using Scipy Optimize
print("Training the model... (This might take a few seconds)")
options = {'maxiter': 100}
lambda_ = 10

res = optimize.minimize(
    fun=lambda x: cofiCostFunc(x, Ynorm, R, num_users, num_movies, num_features, lambda_),
    x0=initial_parameters,
    method='TNC',
    jac=True,
    options=options
)

theta = res.x

# Unfold results
X = theta[:num_movies*num_features].reshape(num_movies, num_features)
Theta = theta[num_movies*num_features:].reshape(num_users, num_features)

print("Training Completed.")
```

## Step 8: Making Recommendations

```python
# 1. Simple function to load movie names
def loadMovieList():
    file = open(EX8MOVIESIDS, encoding='ISO-8859-1')
    movies = file.readlines()
    
    movieNames = []
    for line in movies:
        parts = line.split() 
        name = ' '.join(parts[1:])
        movieNames.append(name)
        
    return movieNames

# 2. Make Predictions
# Calculate predictions: X * Theta (transposed)
p = np.dot(X, Theta.T)

# Add back the mean ratings to get the real score
my_predictions = p[:, 0] + Ymean

# 3. Print Top 10 Recommendations
movieList = loadMovieList()

# Sort predictions from highest to lowest
ix = np.argsort(my_predictions)[::-1]

print('\nTop recommendations for you:')
print('-----------------------------')

for i in range(10):
    j = ix[i] # Get the movie index
    print(f'Predicting rating {my_predictions[j]:.1f} for movie: {movieList[j]}')
```
