# Step 3: K-Means - Initialization

```python
def find_closest_centroids(X, centroids):
    """
    Assign each example in X to the closest centroid.

    :param X: m x n matrix of examples
    :param centroids: K x n matrix of centroids
    :return: idx, vector of centroid assignments (length m)
    """

    m = X.shape[0]
    K = centroids.shape[0]
    idx = np.zeros(m, dtype=int)
    
    for i in range(m):
        distances = np.zeros(K)
        for k in range(K):
            distances[k] = np.sum((X[i] - centroids[k])**2)
        idx[i] = np.argmin(distances)
    
    return idx
```

```python
def compute_centroids(X, idx, K):
    """
    Compute new centroids by averaging the points assigned to each centroid.

    :param X: m x n matrix of examples
    :param idx: vector of centroid assignments
    :param K: number of centroids
    :return: centroids, K x n matrix
    """

    n = X.shape[1]
    centroids = np.zeros((K, n))
    
    for k in range(K):
        points = X[idx == k]
        if len(points) > 0:
            centroids[k] = np.mean(points, axis=0)
    
    return centroids
```

```python
def elbow_plot(X):
    Ks = range(1, 11)
    sse = []

    for k in Ks:
        kmeans = KMeans(n_clusters=k, random_state=42)
        kmeans.fit(X)
        sse.append(kmeans.inertia_)

    plt.plot(Ks, sse, 'o-')
    plt.xlabel("Number of clusters (K)")
    plt.ylabel("SSE (Inertia)")
    plt.title("K-Means Elbow Plot")
    plt.grid(True)
    plt.show()

elbow_plot(X)
```

```python
# Average silhouette score for a range of K
def silhouette_scores_plot(X, k_min=2, k_max=10):
    Ks = range(k_min, k_max + 1)
    scores = []
    for k in Ks:
        km = KMeans(n_clusters=k, n_init=10)
        labels = km.fit_predict(X)
        score = silhouette_score(X, labels)
        scores.append(score)

    plt.figure(figsize=(6,4))
    plt.plot(list(Ks), scores, 'o-')
    plt.xlabel('Number of clusters (K)')
    plt.ylabel('Average silhouette score')
    plt.title('Silhouette scores for different K')
    plt.grid(True)
    plt.show()

# Usage example (run after elbow plot)
silhouette_scores_plot(X, k_min=2, k_max=10)
```

```python
# Simple silhouette bar plot for one chosen K (inspect cluster quality)
def simple_silhouette_plot(X, K):
    km = KMeans(n_clusters=K, random_state=42, n_init=10)
    labels = km.fit_predict(X)
    sil_vals = silhouette_samples(X, labels)   # silhouette value per sample
    avg_score = silhouette_score(X, labels)

    # Prepare plot
    plt.figure(figsize=(8, 4))
    y_lower = 10
    for i in range(K):
        ith_vals = np.sort(sil_vals[labels == i])
        size_cluster_i = ith_vals.shape[0]
        y_upper = y_lower + size_cluster_i

        plt.fill_betweenx(np.arange(y_lower, y_upper),
                          0, ith_vals, alpha=0.7)
        plt.text(-0.05, y_lower + 0.5 * size_cluster_i, str(i))
        y_lower = y_upper + 10  # 10 for spacing between clusters

    plt.axvline(avg_score, color="red", linestyle="--", label=f"avg = {avg_score:.3f}")
    plt.xlabel("Silhouette coefficient")
    plt.ylabel("Samples (cluster groups stacked)")
    plt.title(f"Silhouette plot for K = {K}")
    plt.legend()
    plt.xlim([-0.1, 1])
    plt.show()

# Usage example: inspect chosen K
simple_silhouette_plot(X, K=3)
```

```python
K = 3  # number of clusters
initial_centroids = np.array([[3, 3], [6, 2], [8, 5]])
np.random.seed(42)

# Find closest centroids
idx = find_closest_centroids(X, initial_centroids)
print("Closest centroids for first 3 examples:\n", idx[:3])

# Compute new centroids
centroids = compute_centroids(X, idx, K)
print("Computed centroids:\n", centroids)
```

# Step 4: Run K-Means Clustering

```python
def _compute_cost(X, centroids, idx):
    cost = 0
    for i in range(X.shape[0]):
        cost += np.sum((X[i] - centroids[idx[i]])**2)
    return cost / X.shape[0]


def run_kmeans(X, initial_centroids, max_iters=10, plot_progress=False):
    centroids = initial_centroids.copy()
    K = centroids.shape[0]

    # store history
    history = [centroids.copy()]

    for i in range(max_iters):
        idx = find_closest_centroids(X, centroids)
        centroids = compute_centroids(X, idx, K)
        history.append(centroids.copy())

        if plot_progress:
            plt.figure(figsize=(6,4))

            # scatter points
            plt.scatter(X[:,0], X[:,1], c=idx, cmap='viridis', s=20)

            # plot centroid paths
            H = np.array(history)        # (iterations, K, 2)
            for k in range(K):
                plt.plot(H[:,k,0], H[:,k,1], '-o', linewidth=2)

            # plot current centroids
            plt.scatter(centroids[:,0], centroids[:,1], 
                        c='red', marker='x', s=200, linewidths=3)

            plt.title(f"K-Means iteration {i+1} | Cost: {_compute_cost(X, centroids, idx):.2f}")
            plt.xlabel("X1")
            plt.ylabel("X2")
            plt.show()

    return centroids, idx
```

```python
centroids, idx = run_kmeans(X, initial_centroids, max_iters=3, plot_progress=True)

print("Final centroids:\n", centroids)
```

# Step 5: PCA Example

```python
def feature_normalize(X):
    """Normalize each feature to have mean 0 and std 1."""

    mu = np.mean(X, axis=0)
    sigma = np.std(X, axis=0)
    X_norm = (X - mu) / sigma

    return X_norm, mu, sigma
```

```python
def pca(X):
    m = X.shape[0]
    Sigma = (X.T @ X) / m
    U, S, _ = np.linalg.svd(Sigma)

    return U, np.diag(S)
```

```python
def draw_line(p1, p2, style='-k', linewidth=2):
    plt.plot([p1[0], p2[0]], [p1[1], p2[1]], style, linewidth=linewidth)
```

```python
with open(EX7DATA1, 'rb') as fin:
    X = pickle.load(fin)

# Normalize features
X_norm, mu, sigma = feature_normalize(X)

# Run PCA
U, S = pca(X_norm)

# Visualize top eigenvectors
draw_line(mu, mu + 1.5*S[0,0]*U[:,0].T, style='-k', linewidth=2)
draw_line(mu, mu + 1.5*S[1,1]*U[:,1].T, style='-k', linewidth=2)
plt.scatter(X_norm[:,0], X_norm[:,1], s=20)
plt.title("PCA - Eigenvectors")
plt.show()
```

```python
explained_variance = np.diag(S) / np.sum(np.diag(S))

print("Explained variance ratio:", explained_variance[:5])
```

# Step 6: Dimension Reduction

```python
def project_data(X, U, K):
    U_reduced = U[:, :K]
    return X @ U_reduced
```

```python
def recover_data(Z, U, K):
    U_reduced = U[:, :K]
    return Z @ U_reduced.T
```

```python
K = 1  # project onto 1 dimension
Z = project_data(X_norm, U, K)
X_rec = recover_data(Z, U, K)

print("Original vs Recovered (first example):")
print("Original:", X_norm[0,:])
print("Recovered:", X_rec[0,:])
```

```python
# Visualization
plt.figure(figsize=(6,5))
plt.scatter(X_norm[:, 0], X_norm[:, 1], s=20, label="Original")
plt.scatter(X_rec[:, 0], X_rec[:, 1], s=20, label="Recovered")
plt.title("PCA Dimension Reduction (K=1)")
plt.xlabel("X1")
plt.ylabel("X2")

# Connect original and recovered points for visualization
for i in range(X_norm.shape[0]):
    plt.plot([X_norm[i,0], X_rec[i,0]],
             [X_norm[i,1], X_rec[i,1]], 'k--', linewidth=0.4)

plt.legend()
plt.show()
```

# Step 7: PCA on Faces Dataset

```python
def display_data(X, ax=None):
    """
    Displays 2D data stored in X in a grid.
    X: each row is an example, assumed to be a square image (e.g., 20x20).
    ax: matplotlib Axes object (optional)
    """

    if X.ndim == 1:
        X = X.reshape(1, -1)
    
    m, n = X.shape
    example_width = int(round(math.sqrt(n)))
    example_height = n // example_width
    
    # Compute rows/cols for display grid
    display_rows = int(math.floor(math.sqrt(m)))
    display_cols = int(math.ceil(m / display_rows))
    
    # Padding between images
    pad = 1
    display_array = -np.ones((pad + display_rows * (example_height + pad),
                               pad + display_cols * (example_width + pad)))
    
    curr_ex = 0
    for i in range(display_rows):
        for j in range(display_cols):
            if curr_ex >= m:
                break
            # Copy image
            top = pad + i * (example_height + pad)
            left = pad + j * (example_width + pad)
            img = X[curr_ex].reshape(example_height, example_width)
            display_array[top:top+example_height, left:left+example_width] = img
            curr_ex += 1
    
    if ax is None:
        plt.imshow(display_array, cmap='gray')
        plt.axis('off')
    else:
        ax.imshow(display_array, cmap='gray')
        ax.axis('off')
```

```python
with open(EX7FACES, 'rb') as fin:
    X_faces = pickle.load(fin)

# Normalize
X_faces_norm, mu, sigma = feature_normalize(X_faces)

# Run PCA
U, S = pca(X_faces_norm)

# Display top 36 eigenfaces
display_data(U[:, 0:36].T)
plt.title("Top 36 Eigenfaces")
plt.show()

# Project onto K=100 dimensions
K = 100
Z_faces = project_data(X_faces_norm, U, K)
X_rec_faces = recover_data(Z_faces, U, K)

# Compare original vs reconstructed
fig, (ax0, ax1) = plt.subplots(1, 2, figsize=(12,6))

display_data(X_faces_norm[0:100,:], ax=ax0)
ax0.set_title("Original Faces")

display_data(X_rec_faces[0:100,:], ax=ax1)
ax1.set_title("Recovered Faces")

plt.show()
```

# Step 8: K-Means Image Compression

```python
def kmeans_init_centroids(X, K):
    """
    Initialize K centroids by randomly sampling from the dataset X.
    """

    m = X.shape[0]
    indices = np.random.choice(m, K, replace=False)

    return X[indices]
```

```python
# Load image
A = plt.imread(EX7SMALLBIRD)
img_size = A.shape
X_img = np.reshape(A, (img_size[0]*img_size[1], 3))

# Run K-Means
K = 16
max_iters = 10
initial_centroids = kmeans_init_centroids(X_img, K)
centroids, idx = run_kmeans(X_img, initial_centroids, max_iters)

# Compress image
X_recovered = centroids[idx, :]
X_recovered = np.reshape(X_recovered, (img_size[0], img_size[1], 3))

# Show original vs compressed
fig, (ax0, ax1) = plt.subplots(1,2)

ax0.imshow(A)
ax0.set_title("Original")

ax1.imshow(X_recovered)
ax1.set_title(f"Compressed with {K} colors")

plt.show()
```
