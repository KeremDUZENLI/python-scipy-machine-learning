# Step 1: Helper Functions

## Step 1.1: Email Helpers

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import re
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import GridSearchCV
from nltk.stem.porter import PorterStemmer
```

```python
def read_text_file(path):
    with open(path, 'r', encoding='utf-8', errors='ignore') as f:
        return f.read()
```

```python
def get_vocab_list(vocab_path):
    vocab = {}

    with open(vocab_path, 'r', encoding='utf-8', errors='ignore') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 2:
                idx = int(parts[0])
                word = parts[1]
                vocab[word] = idx

    return vocab
```

```python
porter = PorterStemmer()

def process_email(email_contents, verbose=False):
    """Return list of word indices (1-based indices like original exercise)."""
    
    # Lowercase
    email_contents = email_contents.lower()
    
    # Remove HTML
    email_contents = re.sub('<[^<>]+>', ' ', email_contents)
    
    # Replace numbers, urls, email addr, dollar sign
    email_contents = re.sub('[0-9]+', 'number', email_contents)
    email_contents = re.sub('(http|https)://[^\s]*', 'httpaddr', email_contents)
    email_contents = re.sub('[^\s]+@[^\s]+', 'emailaddr', email_contents)
    email_contents = re.sub('[$]+', 'dollar', email_contents)
    
    # Tokenize
    tokens = re.split(r'[@$/#\.\-:&\*\+=\[\]?!(){},\'\">_<;%\s\n\r\t]+', email_contents)
    word_indices = []
    vocab = get_vocab_list(EX6VOCAB)
    for token in tokens:
        token = re.sub('[^a-zA-Z0-9]', '', token)
        token = token.strip()
        if len(token) == 0:
            continue
        stemmed = porter.stem(token)
        if stemmed in vocab:
            word_indices.append(vocab[stemmed])  # vocab indices are 1-based
        if verbose:
            print(stemmed, end=' ')
    if verbose:
        print()

    return word_indices
```

```python
def email_features(word_indices, n=1899):
    """Convert word_indices (list of ints) to binary feature vector (n x 1)."""
    x = np.zeros((n, 1), dtype=int)
    for idx in set(word_indices):
        if 1 <= idx <= n:
            x[idx-1, 0] = 1
            
    return x
```

## Step 1.2: SVM Helpers

```python
def load_mat(path):
    """Load a MATLAB .mat file and return its dict."""
    return scipy.io.loadmat(path)
```

```python
def plot_data(X, y, title='Data'):
    pos = np.where(y.flatten() == 1)[0]
    neg = np.where(y.flatten() == 0)[0]
    
    plt.figure(figsize=(6,5))
    plt.plot(X[pos, 0], X[pos, 1], 'k+', linewidth=1, markersize=7, label='Positive (1)')
    plt.plot(X[neg, 0], X[neg, 1], 'yo', markerfacecolor='y', markersize=7, label='Negative (0)')
    
    plt.title(title)
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')
    plt.legend()
    plt.grid(True)
    
    plt.show()
```

```python
def gaussian_kernel(x1, x2, sigma):
    """Return Gaussian kernel value between two 1D arrays (vectors)."""
    x1 = np.asarray(x1).ravel()
    x2 = np.asarray(x2).ravel()
    diff = x1 - x2
    
    return np.exp(-np.dot(diff, diff) / (2 * (sigma ** 2)))
```

```python
def sigma_to_gamma(sigma):
    """Convert sigma to sklearn's gamma for RBF kernel: gamma = 1/(2*sigma^2)"""
    return 1.0 / (2.0 * (sigma ** 2))
```

```python
def visualize_boundary(X, y, model, title='Decision boundary', grid_size=100):
    # X: (m,2)  model: sklearn-like with predict()
    x1min, x1max = X[:,0].min(), X[:,0].max()
    x2min, x2max = X[:,1].min(), X[:,1].max()
    
    x1 = np.linspace(x1min - 0.5, x1max + 0.5, grid_size)
    x2 = np.linspace(x2min - 0.5, x2max + 0.5, grid_size)
    
    X1, X2 = np.meshgrid(x1, x2)
    grid = np.c_[X1.ravel(), X2.ravel()]
    Z = model.predict(grid).reshape(X1.shape)
    
    plt.figure(figsize=(6,5))
    plt.contour(X1, X2, Z, levels=[0.5], linewidths=1, colors='blue')
    plot_data(X, y, title=title)
```

```python
def dataset_3_params(X, y, Xval, yval, C_list=None, sigma_list=None):
    if C_list is None:
        C_list = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    if sigma_list is None:
        sigma_list = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]

    best = {'C': None, 'sigma': None, 'acc': -1}
    for C in C_list:
        for sigma in sigma_list:
            gamma = sigma_to_gamma(sigma)
            clf = make_pipeline(StandardScaler(), SVC(C=C, kernel='rbf', gamma=gamma))
            clf.fit(X, y.ravel())
            acc = clf.score(Xval, yval.ravel())
            if acc > best['acc']:
                best = {'C': C, 'sigma': sigma, 'acc': acc}

    return best['C'], best['sigma'], best['acc']
```

## Step 2.3: Train Linear SVM

```python
# Load spam train & test
data = load_mat(EX6SPAMTRAIN)
X_train, y_train = data['X'], data['y']

data = load_mat(EX6SPAMTEST)
X_test, y_test = data['Xtest'], data['ytest']

# Train linear SVM
clf = make_pipeline(StandardScaler(), SVC(C=0.1, kernel='linear', max_iter=10000))
clf.fit(X_train, y_train.ravel())

# Evaluate
train_acc = clf.score(X_train, y_train.ravel()) * 100
test_acc = clf.score(X_test, y_test.ravel()) * 100

print(f"Training Accuracy: {train_acc:.2f}%")
print(f"Test Accuracy: {test_acc:.2f}%")
```

## Step 2.4: Print Spam Words

```python
# Top predictors (linear SVM weights)
sv = clf.named_steps['svc']
weights = sv.coef_[0]  # shape (n_features,)

# Get vocab list mapping index->word for printing
vocab_map = {v:k for k,v in get_vocab_list(EX6VOCAB).items()}  # reversed: idx->word
top_idx = np.argsort(weights)[-15:][::-1]

print("\nTop predictors of spam:")
for i in top_idx:
    word = vocab_map.get(i+1, "N/A")
    print(f"{word:15} ({weights[i]:.4f})")
```

## Step 3.2: Dataset 2 (RBF)

```python
data = load_mat(EX6DATA2)
X2, y2 = data['X'], data['y']
plot_data(X2, y2, title='Dataset 2')

sigma = 0.1
gamma = sigma_to_gamma(sigma)

clf_rbf = make_pipeline(StandardScaler(), SVC(C=1.0, kernel='rbf', gamma=gamma))
clf_rbf.fit(X2, y2.ravel())

visualize_boundary(X2, y2, clf_rbf, title=f'RBF SVM (Dataset 2), sigma={sigma}')
```

## Step 3.3: Dataset 3 (use cross-val to pick C and sigma)

```python
data = load_mat(EX6DATA3)
X3, X3val, y3, y3val = data['X'], data['Xval'], data['y'], data['yval']
C_best, sigma_best, acc = dataset_3_params(X3, y3, X3val, y3val)

print("Best C, sigma found:", C_best, sigma_best, "val-acc:", acc)

gamma_best = sigma_to_gamma(sigma_best)
clf_best = make_pipeline(StandardScaler(), SVC(C=C_best, kernel='rbf', gamma=gamma_best))
clf_best.fit(X3, y3.ravel())

visualize_boundary(X3, y3, clf_best, title=f'Dataset 3 boundary (C={C_best}, sigma={sigma_best})')
```
