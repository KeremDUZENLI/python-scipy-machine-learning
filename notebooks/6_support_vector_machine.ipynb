# Step 1: Import Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.io
import re
from sklearn.svm import SVC
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from nltk.stem.porter import PorterStemmer
```

# Step 2: Email

```python
def read_text_file(path):
    with open(path, 'r', encoding='utf-8', errors='ignore') as f:
        return f.read()
```

```python
def get_vocab_list(vocab_path):
    vocab = {}

    with open(vocab_path, 'r', encoding='utf-8', errors='ignore') as f:
        for line in f:
            parts = line.strip().split()
            if len(parts) >= 2:
                idx = int(parts[0])
                word = parts[1]
                vocab[word] = idx

    return vocab
```

```python
def process_email(email_contents):
    vocab_list = get_vocab_list(EX6VOCAB)
    word_indices = []
    
    email_contents = email_contents.lower()
    email_contents = re.sub("<[^<>]+>", " ", email_contents)
    email_contents = re.sub("[0-9]+", "number", email_contents)
    email_contents = re.sub("(http|https)://[^\s]+", "httpaddr", email_contents)
    email_contents = re.sub("[^\s]+@[^\s]+", "emailaddr", email_contents)
    email_contents = re.sub("[$]+", "dollar", email_contents)

    stemmer = PorterStemmer()
    tokens = re.split(r'[@$/#.-:&\*\+=\[\]?!(){},\'\'\">_<;%\s\n\r\t]+', email_contents)

    for s in tokens:
        s = re.sub("[^a-zA-Z0-9]", "", s)
        s = stemmer.stem(s.strip())
        if len(s) < 1:
            continue
        for idx, word in vocab_list.items():
            if word == s:
                word_indices.append(idx)
                break

    return word_indices
```

```python
def email_features(word_indices):
    n = 1899
    x = np.zeros((n, 1))

    for idx in word_indices:
        x[idx - 1] = 1  # Adjust for 0-based indexing

    return x
```

```python
def gaussian_kernel(x1, x2, sigma):
    """Works for 1-D arrays or lists; returns a scalar float."""
    
    x1 = np.asarray(x1).reshape(-1)
    x2 = np.asarray(x2).reshape(-1)
    diff = x1 - x2
    norm2 = np.dot(diff, diff)
    sim = np.exp(-norm2 / (2.0 * sigma**2))

    return float(sim)
```

## Step 2.3: Train Linear SVM

```python
# Load spam train & test
data = scipy.io.loadmat(EX6SPAMTRAIN)
X_train, y_train = data['X'], data['y']

data = scipy.io.loadmat(EX6SPAMTEST)
X_test, y_test = data['Xtest'], data['ytest']

# Train linear SVM
clf = make_pipeline(StandardScaler(), SVC(C=0.1, kernel='linear', max_iter=10000))
clf.fit(X_train, y_train.ravel())

# Evaluate
train_acc = clf.score(X_train, y_train.ravel()) * 100
test_acc = clf.score(X_test, y_test.ravel()) * 100

print(f"Training Accuracy: {train_acc:.2f}%")
print(f"Test Accuracy: {test_acc:.2f}%")
```

## Step 2.4: Print Spam Words

```python
# Top predictors (linear SVM weights)
sv = clf.named_steps['svc']
weights = sv.coef_[0]  # shape (n_features,)

# Get vocab list mapping index->word for printing
vocab_map = {v:k for k,v in get_vocab_list(EX6VOCAB).items()}  # reversed: idx->word
top_idx = np.argsort(weights)[-15:][::-1]

print("\nTop predictors of spam:")
for i in top_idx:
    word = vocab_map.get(i+1, "N/A")
    print(f"{word:15} ({weights[i]:.4f})")
```

# Step 3: SVM

```python
def plot_data(X, y, ax=None, title='Data'):
    """
    Plot 2D data points with positive (1) and negative (0) markers.
    """
    if ax is None:
        fig, ax = plt.subplots(figsize=(6,5))

    pos = np.where(y.flatten() == 1)[0]
    neg = np.where(y.flatten() == 0)[0]

    ax.plot(X[pos, 0], X[pos, 1], 'k+', linewidth=1, markersize=7, label='Positive (1)')
    ax.plot(X[neg, 0], X[neg, 1], 'yo', markerfacecolor='y', markersize=7, label='Negative (0)')

    ax.set_xlabel('Feature 1')
    ax.set_ylabel('Feature 2')
    ax.set_title(title)
    ax.legend()
    ax.grid(True)

    return ax
```

## Step 3.1: Dataset 1 (linear SVM)

```python
def visualize_boundary(X, y, model, title='Decision boundary', grid_size=100):
    x1min, x1max = X[:,0].min(), X[:,0].max()
    x2min, x2max = X[:,1].min(), X[:,1].max()
    
    x1 = np.linspace(x1min - 0.5, x1max + 0.5, grid_size)
    x2 = np.linspace(x2min - 0.5, x2max + 0.5, grid_size)
    
    X1, X2 = np.meshgrid(x1, x2)
    grid = np.c_[X1.ravel(), X2.ravel()]
    Z = model.predict(grid).reshape(X1.shape)
    
    plt.figure(figsize=(6,5))
    ax = plt.gca()  # get current axes
    ax.contour(X1, X2, Z, levels=[0.5], linewidths=1, colors='blue')
    plot_data(X, y, ax=ax, title=title)
    ax.set_title(title)

    plt.show()
```

```python
data = scipy.io.loadmat(EX6DATA1)
X, y = data['X'], data['y']
plot_data(X, y, title='Dataset 1')

clf_lin = make_pipeline(StandardScaler(), SVC(C=1.0, kernel='linear'))
clf_lin.fit(X, y.ravel())
visualize_boundary(X, y, clf_lin, title='Linear SVM (Dataset 1)')
```

## Step 3.2: Dataset 2 (RBF)

```python
def sigma_to_gamma(sigma):
    """Convert sigma to sklearn's gamma for RBF kernel: gamma = 1/(2*sigma^2)"""
    return 1.0 / (2.0 * (sigma ** 2))
```

```python
data = scipy.io.loadmat(EX6DATA2)
X2, y2 = data['X'], data['y']
plot_data(X2, y2, title='Dataset 2')

sigma = 0.1
gamma = sigma_to_gamma(sigma)

clf_rbf = make_pipeline(StandardScaler(), SVC(C=1.0, kernel='rbf', gamma=gamma))
clf_rbf.fit(X2, y2.ravel())

visualize_boundary(X2, y2, clf_rbf, title=f'RBF SVM (Dataset 2), sigma={sigma}')
```

## Step 3.3: Dataset 3 (use cross-val to pick C and sigma)

```python
def svm_train(X, y, C, sigma):
    gamma = 1 / (2 * sigma**2)
    clf = SVC(C=C, kernel='rbf', gamma=gamma)
    clf.fit(X, y.ravel())
    return clf


def svm_predict(model, X):
    return model.predict(X)


def dataset_3_params(X, y, Xval, yval):
    C_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    sigma_values = C_values

    best_error = float('inf')
    best_C = C_values[0]
    best_sigma = sigma_values[0]

    for C in C_values:
        for sigma in sigma_values:
            model = svm_train(X, y, C, sigma)
            predictions = svm_predict(model, Xval)
            error = np.mean(predictions != yval)
            if error < best_error:
                best_error = error
                best_C = C
                best_sigma = sigma

    return best_C, best_sigma
```

```python
# Load data
data = scipy.io.loadmat(EX6DATA3)
X3, X3val, y3, y3val = data['X'], data['Xval'], data['y'], data['yval']

# Find best C and sigma
C_best, sigma_best = dataset_3_params(X3, y3, X3val, y3val)

# Compute validation accuracy
model = svm_train(X3, y3, C_best, sigma_best)
model_predictions = svm_predict(model, X3val)
acc = np.mean(model_predictions == y3val.ravel())

print("Best C, sigma found:", C_best, sigma_best, "val-acc:", acc)

# Train scikit-learn SVM for visualization
gamma_best = sigma_to_gamma(sigma_best)
clf_best = make_pipeline(StandardScaler(), SVC(C=C_best, kernel='rbf', gamma=gamma_best))
clf_best.fit(X3, y3.ravel())

visualize_boundary(X3, y3, clf_best, title=f'Dataset 3 boundary (C={C_best}, sigma={sigma_best})')
```

## Step 3.4: Gaussian

```python
def gaussian_kernel(x1, x2, sigma):
    """
    Gaussian / RBF kernel: sim = exp(-||x1 - x2||^2 / (2*sigma^2))
    """
    x1 = np.asarray(x1).ravel()
    x2 = np.asarray(x2).ravel()

    diff = x1 - x2
    sim = np.exp(-(diff @ diff) / (2 * sigma**2))

    return sim

  
def compute_kernel_matrix(X1, X2, sigma):
    K = np.zeros((X1.shape[0], X2.shape[0]))
    
    for i in range(X1.shape[0]):
        for j in range(X2.shape[0]):
            K[i, j] = gaussian_kernel(X1[i], X2[j], sigma)
    
    return K
```

```python
def svm_train_custom_kernel(X, y, C, sigma):
    K = compute_kernel_matrix(X, X, sigma)
    clf = SVC(C=C, kernel='precomputed')
    clf.fit(K, y.ravel())
  
    return clf


def svm_predict_custom_kernel(model, X_train, X_new, sigma):
    K_test = compute_kernel_matrix(X_new, X_train, sigma)
  
    return model.predict(K_test)


def dataset_3_params_custom(X, y, Xval, yval):
    C_values = [0.01, 0.03, 0.1, 0.3, 1, 3, 10, 30]
    sigma_values = C_values
    
    best_error = float('inf')
    best_C = None
    best_sigma = None

    for C in C_values:
        for sigma in sigma_values:
            model = svm_train_custom_kernel(X, y, C, sigma)
            preds = svm_predict_custom_kernel(model, X, Xval, sigma)
            error = np.mean(preds != yval.ravel())
            if error < best_error:
                best_error = error
                best_C = C
                best_sigma = sigma
    
    return best_C, best_sigma
```

```python
def visualize_boundary_custom(X, y, model, X_train, sigma, title='Decision boundary', grid_size=100):
    """
    Visualize decision boundary for 2D data using a custom Gaussian-kernel SVM.
    """
    x1min, x1max = X[:,0].min(), X[:,0].max()
    x2min, x2max = X[:,1].min(), X[:,1].max()

    x1 = np.linspace(x1min - 0.5, x1max + 0.5, grid_size)
    x2 = np.linspace(x2min - 0.5, x2max + 0.5, grid_size)

    X1, X2 = np.meshgrid(x1, x2)
    grid = np.c_[X1.ravel(), X2.ravel()]

    # Use custom kernel prediction
    Z = svm_predict_custom_kernel(model, X_train, grid, sigma).reshape(X1.shape)

    plt.figure(figsize=(6,5))
    ax = plt.gca()  # current axes
    ax.contour(X1, X2, Z, levels=[0.5], linewidths=1, colors='blue')
    plot_data(X, y, ax=ax, title=title)
    
    plt.show()
```

```python
data = scipy.io.loadmat(EX6DATA3)
X3, X3val, y3, y3val = data['X'], data['Xval'], data['y'], data['yval']

# best params
C_best, sigma_best = dataset_3_params_custom(X3, y3, X3val, y3val)

# compute accuracy
model = svm_train_custom_kernel(X3, y3, C_best, sigma_best)
preds = svm_predict_custom_kernel(model, X3, X3val, sigma_best)
acc = np.mean(preds == y3val.ravel())

print("Best C:", C_best, "Best sigma:", sigma_best, "Validation accuracy:", acc)

# Visualize decision boundary
visualize_boundary_custom(X3, y3, model, X3, sigma_best, title=f'Dataset 3 (C={C_best}, sigma={sigma_best})')
```
